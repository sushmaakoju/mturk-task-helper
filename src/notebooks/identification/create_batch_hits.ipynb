## Now for each row in csv, create HITLayoutParameters  
### `[{'Name': colname, "Value": row_dict[colname]} for colname in colnames[:20] ] ` 
### is of format  
#### `[{'Name': image1_url, "Value": 'https://mturk-s3-cg.s3.amazonaws.com/task2/selwyn/smallbatch/0001_0001/0001-0001-12960-4800.png'},{'Name': image3_url, "Value": 'https://mturk-s3-cg.s3.amazonaws.com/task2-images/LINZ/0004_0001/0004_0001_10899-12065-small.png'} ]  `

```python
HITLayoutParameters = []
hits = []
for row_dict in imageurl_df.to_dict(orient='records'):
    HITLayoutParameters = [{'Name': colname, "Value": row_dict[colname]} for colname in colnames[:20] ]
    
    #print(HITLayoutParameters)
    result = create_hit(HITLayoutParameters)
    assert result, "HIT creation failed. Please check and try again."
    hits.append(result)  
    #print([{'Name': colname, "value": row_dict[colname]} for colname in colnames ])
```
